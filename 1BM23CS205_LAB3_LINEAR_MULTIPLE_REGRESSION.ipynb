```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("canada_per_capita_income.csv")

# Feature and target
X = df[['year']]
y = df['per capita income (US$)']

# Train model
model = LinearRegression()
model.fit(X, y)

# Prediction for 2020
income_2020 = model.predict([[2020]])

print("Predicted per capita income for Canada in 2020:",
      income_2020[0])

# Visualization (optional but recommended for record)
plt.scatter(X, y)
plt.plot(X, model.predict(X))
plt.xlabel("Year")
plt.ylabel("Per Capita Income")
plt.title("Canada Income Trend")
plt.show()
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("salary.csv")

# Drop rows with NaN values
df.dropna(inplace=True)

X = df[['YearsExperience']]
y = df['Salary']

model = LinearRegression()
model.fit(X, y)

# Predict salary for 12 years experience
salary_12 = model.predict([[12]])

print("Predicted salary for 12 years experience:",
      salary_12[0])

# Plot
plt.scatter(X, y)
plt.plot(X, model.predict(X))
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.title("Salary vs Experience")
plt.show()
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# Load dataset
df = pd.read_csv("hiring.csv")

# Fill missing values if present
df = df.fillna(0)

# Convert 'experience' column from word to number
# Known values from inspecting the dataframe:
# 'five': 5, 'two': 2, 'seven': 7, 'three': 3, 'ten': 10, 'eleven': 11
word_to_num = {
    'one': 1, 'two': 2, 'three': 3, 'four': 4, 'five': 5,
    'six': 6, 'seven': 7, 'eight': 8, 'nine': 9, 'ten': 10,
    'eleven': 11, 'twelve': 12, 'zero': 0
}
df['experience'] = df['experience'].apply(lambda x: word_to_num.get(str(x).lower(), x))
df['experience'] = pd.to_numeric(df['experience'], errors='coerce').fillna(0)

X = df[['experience', 'test_score(out of 10)', 'interview_score(out of 10)']]
y = df['salary($)']

model = LinearRegression()
model.fit(X, y)

# Predictions required by lab
pred1 = model.predict([[2, 9, 6]])
pred2 = model.predict([[12, 10, 10]])

print("Salary for (2 yr, 9 score, 6 interview):", pred1[0])
print("Salary for (12 yr, 10 score, 10 interview):", pred2[0])
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# Load dataset
df = pd.read_csv("1000_Companies.csv")

# One-Hot Encoding for 'State'
df = pd.get_dummies(df, columns=['State'], drop_first=True)

# Features and target
X = df.drop('Profit', axis=1)
y = df['Profit']

model = LinearRegression()
model.fit(X, y)

# Prediction for given values
new_company = [[91694.48, 515841.3, 11931.24, 0, 1]]
# (Example assumes Florida encoded as last column)

profit_pred = model.predict(new_company)

print("Predicted profit:", profit_pred[0])
```
